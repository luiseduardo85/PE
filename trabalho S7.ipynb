# Questão 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/weight_chart.txt', delimiter='\t')
plt.figure(figsize=(10,8))
plt.title("The relationship between age and weight in a growing infant",fontweight="bold")
plt.plot(data.Age,data.Weight,color='blue', ls='-', lw='2', marker='o',ms=15, markeredgecolor='white',markeredgewidth=3)

plt.ylabel('Weight (kg)', fontsize='16')
plt.xlabel('Age (months)',fontsize='16')

plt.show()
```

# Questão 3

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

amostra = np.random.standard_normal(size=10000)

amostra2 = np.random.standard_normal(size=10000)

dados = []

for amostras in amostra2:
  dados.append(amostras+4)


amostras_concatenada = np.concatenate([amostra, dados])

plt.figure(figsize=(8,10))
plt.hist(amostras_concatenada, bins=100, color='blue', ec='yellow')
plt.ylabel('Frequencia', fontsize='14')
plt.xlabel('Valores', fontsize='14')


plt.show()
```

# Questão 5

```python
data = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/up_down_expression.txt', delimiter='\t')

meio = data.loc[data['State']=='unchanging']
superior = data.loc[data['State']=='up']
inferior = data.loc[data['State']=='down']

plt.figure(figsize=(8,10))
plt.scatter(meio.Condition1, meio.Condition2, color='gray',s=5)
plt.scatter(superior.Condition1, superior.Condition2, color='red',s=5)
plt.scatter(inferior.Condition1,inferior.Condition2, color='blue',s=5)
plt.xlabel('Condição 1', fontsize='14')
plt.ylabel('Condição 2', fontsize='14')

plt.show()
```

# Questão 6

```python
data = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/chromosome_position_data.txt', delimiter='\t')


plt.figure(figsize=(8,8))
plt.plot(data.Mut1, color='red')

plt.plot(data.Mut2, color='blue')

plt.plot(data.WT, color='green')

plt.plot(data.Mut1,'-r',label='Mut1')

plt.plot(data.Mut2,'-b',label='Mut2')

plt.plot(data.WT,'-g',label='WT')

plt.legend(loc='upper left',)

plt.ylabel('Valor', fontsize='16')
plt.xlabel('Posição sem cromossomos', fontsize='16')

plt.show()
```

# Questão 7

```python
data = pd.read_csv('https://raw.githubusercontent.com/androidpcguy/bggn213/master/class05/bimm143_05_rstats/brain_bodyweight.txt', delimiter='\t')

plt.figure(figsize=(8,10))
for elemento in data.values:
  name = elemento[0]
  n = elemento[1]
  m = elemento[2]
  xbar = elemento[3]
  ybar = elemento[4]

  plt.errorbar(x=n, y=m, xerr=xbar, yerr=ybar, color='black', marker='h' )
  plt.annotate(name, (n-0.13,m-0.2), fontsize='small')


plt.ylabel('Brain Weight', fontsize='14')
plt.xlabel('Body Weight', fontsize='14')
plt.show()
```
